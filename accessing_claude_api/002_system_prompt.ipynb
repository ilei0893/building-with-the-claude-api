```python
# Create an API client

from anthropic import Anthropic

client = Anthropic()

model = "claude-sonnet-4-0"
```

```python
def add_user_message(messages, text):
    user_message = {
            "role" : "user",
            "content": text
        }
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {
            "role" : "assistant",
            "content": text
        }
    messages.append(assistant_message)

def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages
    }

    if system:
        params["system"] = system

    message = client.messages.create(**params)

    return message.content[0].text
```

```python
import json

messages = []

add_user_message(messages,
                 "Write a Python function that checks a string for duplicate characters.")

system = """
You are an expert coder. Implement functions as concisely as possible without explaining what you're doing.
"""
answer = chat(messages, system)

lines = answer.strip('`').strip().splitlines()
code_lines = lines[1:]  # skip the 'python' language tag
code = '\n'.join(code_lines)
code
```
